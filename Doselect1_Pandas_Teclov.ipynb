> Cleaning mapping dataframe in which "category list" column consist of "0" in some string, replace the "0" from "Na" and "na".

```python
def mapping0(category):
    if '0' in category:
        if category.startswith("0"):    
            return category[:category.find('0')]+'Na'+category[category.find('0')+1:]
        elif category.endswith("0"):  
            return category
        else:
            return category[:category.find('0')]+'na'+category[category.find('0')+1:]
    else:
        return category
```

```python
mapping['category_list']=list(map(mapping0,mapping['category_list']))
```

```python
mapping = mapping.drop('Company_name_new',axis=1)
```

```python
mainsec = pd.melt(mapping, id_vars=['category_list'], var_name=['main_sector'])
mainsec = mainsec[mainsec.value==1]
mainsec = mainsec.drop('value',axis=1)
mainsec.head()
```
